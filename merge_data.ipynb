```python
def read_wav(row):
    """Reads a wav file and assigns the frequency and signal to the dataframe row.
    Args:
        row (pandas.Series): A row of the dataframe containing the filename of the wav file.

    Returns:
        pandas.Series: The updated row with the frequency and signal assigned.

    """
    filename = FILE_DIR+row['filename']
    frequency, signal = wavfile.read(filename)
    row['frequency'] = frequency
    row['signal'] = signal
    return row
```

```python
# TEMP
df_test = df_train_index.head(100)
# Check funcion
# Apply the function to each row in the dataframe
df_test = df_test.progress_apply(read_wav, axis=1)

# Print the updated dataframe
df_test
```
